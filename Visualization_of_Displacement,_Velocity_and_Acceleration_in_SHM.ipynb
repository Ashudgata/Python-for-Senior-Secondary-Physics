```python
#Wave Expression x= a*sin(wt+p_o)
import numpy as np
import matplotlib.pyplot as plt

# Parameters
frequency = 1  # Frequency in Hz
amplitude = 1  # Amplitude of the sine wave
duration = 2   # Duration in seconds
initial_phase = 0  #Initial phase in radians
sampling_rate = 50000 # Sampling rate in Hz

# Time vector from 0 to duration with the given sampling rate
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate sine wave
Displacement = amplitude * np.sin((2 * np.pi * frequency * t) + initial_phase)

# Plot the sine wave
plt.figure(figsize=(16,9))
plt.plot(t, Displacement)
plt.title(f'Displacement in SHM at {frequency} Hz')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
Velocity = np.gradient(Displacement,t)
plt.figure(figsize=(16,9))
plt.plot(t, Velocity)
plt.title(f'Velocity in SHM at {frequency} Hz')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
Acceleration = np.gradient(Velocity,t)
plt.figure(figsize=(16,9))
plt.plot(t, Acceleration)
plt.title(f'Acceleration in SHM at {frequency} Hz')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Visualizing on a single graph
plt.figure(figsize=(16,9))
plt.plot(t, Displacement, label = 'Displacement(m)')
plt.plot(t,Velocity, label = 'Velocity(m/s)')
plt.plot(t,Acceleration, label = 'Acceleration(m/s^2)')
plt.title(f'SHM at {frequency} Hz')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
